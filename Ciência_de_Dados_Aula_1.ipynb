# Criando dataframes

> **Dataframe:** <div>
Antes de partir para visualização e acesso aos dados, é importante definir uma palavra que será muito utilizada no processo de análise de dados.<div>
Um **Dataframe** é tipo bidimensional, que diferente de uma matriz, pode conter diversos tipos de dado, é uma espécie de tabela ou uma planilha da sua base de dado, no qual, cada coluna representa uma variável e cada linha um conjunto de observações, um registro.<div>
Por exemplo, num cadastro de usuários, deseja-se armazenar nome, idade e e-mail. Cada coluna deve representar uma variável, e cada linha um registro, ou seja, cada linha representa um usuário.<div>
Além de importados, como demonstrado acima, os dataframes também podem ser criados manualmente.<div>
O Pandas é a bilbioteca responsável pela manipualção desses dataframes, facilitando a utilização dos dados.



Para criação utilização o método **DataFrame** do Pandas.<div>

Deve-se colocar colchetes para iniciar e encerrar a entrada de dados. Uma coluna deve conter um nome, escrito entre aspas, seguido de dois pontos. Logo após, os dados daquela coluna precisam estar separados entre vírgulas. Colchetes abrem e encerram os dados de uma coluna. Para mover para próxima, utiliza-se vírgula para separar.


```
exemplo = pd.DataFrame({'Nome' : ["Bruna", "Laio", "William"],
                        'idade' : [22, 21, 21], 
                        'Curso' : ["Eng Quimica", "Eng Quimica", "Eng Quimica"]})


exemplo
```

```
exemplo1 = pd.DataFrame({'Nome' : ["John", "Lu", "Zé"],
                        'idade' : [22, 23, 25], 
                        'Curso' : ["Eng Quimica", "Eng Quimica", "Eng Controle"]})

exemplo1
```

```
teste = pd.concat([exemplo, exemplo1],ignore_index= True)
teste
```

# **Indexação e acesso**

> **As colunas do dataframe**<div>
Por vezes, faz-se necessário acessar determinada coluna do dataframe. Existem algumas formas de fazer isso. Antes de apresentar essas maneiras é importante dizer que uma coluna é chamdada de **série**.

> **Série:**<div>
Uma série é um array/lista de valores de um mesmo tipo. As séries possuem índices que identificam as linhas, e estão dispostos do lado esquerdo. Como é comum em programação, o primeiro índice é zero e, nesse caso, não inclui o título daquela coluna.


> ## **Acessando colunas do dataframe**<div>
Para acessar uma coluna, podemos usar a forma mais próxima do que há em orientação objeto. Entretanto, isso só é possível se não há separação entre as palavras ou barras, pontos e outros caracteres especiais, com exceção do underline.

**Utilização:** <div>

**nome_do_dataframe.nome_da_coluna**

Exemplo:
     dados.Total

```
dados.Posição 
```

Vejamos o que acontece ao tentar utilizar essa forma com a coluna 2008/Abr, que possui uma barra no nome:

Dessa maneira, escreve-se o nome do dataframe seguido do nome da coluna, entre chaves e aspas.

## **Filtros ou Seleção condicional**<div>

Com o Pandas, podemos facilmente obter recortes dos dados e filtrar apenas a informação necessárias. Vejamos algumas formas de fazer isso:

```
dados.head()
```

### **loc** <div>
.loc permite o acesso de um grupo de linhas ou colunas através de um label ou de um vetor boleano.
**df.loc[<linhas>, <colunas>]**

Exemplo:
     dados.loc[dados.idade > 23]

### **iloc**

Funciona de maneira semelhante, mas o indexador são números inteiros referente aos indices do registro.

*Quando nenhum item é encontrado pelo loc, ele retorna um KeyError.*


```
dados["Peso (kg)"]>70
```

```
dados.loc[dados["Peso (kg)"]>70]#["Nome"]
```

Operador lógico "&" 

```
dados.loc[(dados["Peso (kg)"]>70) & (dados["Altura (cm)"]>190)]
```

Operador lógico "ou" = |

```
dados.loc[(dados["Peso (kg)"]>70) | (dados["Altura (cm)"]>190)]
```

## **Média de colunas**

Para calcular a média de uma coluna sem o Pandas, seria necessário varrer linha a linha com alguma estrutura de repetição e somar cada um dos valores e em seguida, dividir pelo total de colunas. Com o Pandas basta utilizar o método **mean**, que permite não só obter a média de uma coluna, mas de várias ou até de todas as colunas do dataframe que possuem valores numéricos.


**Utilização:** <div>

1:  
**nome_do_dataframe[["nome_da_coluna1", "nome_da_coluna2"]].mean()**

2:  
**lista = ["nome_da_coluna1", "nome_da_coluna2"]** 

**nome_do_dataframe[lista].mean()**

3:

**nome_do_dataframe.mean()**



Exemplo:
     dados["Total"].mean()

> *Uma observação importante sobre valores númericos:<div> É possível que valores visualmente númericos podem ser strings, ou seja, letras, e isso se dá por uma série de motivos, um deles foi abordado no [notebook anterior](https://colab.research.google.com/drive/1Cd_udAOr30z2Ah6JKeUT6t3xGcV-JWlg?usp=sharing). Uma outra dica para quando isso ocorrer, é olhar a função [.to_numeric](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html) do Pandas.*




```
#Calculando a média de um mês
dados["Peso (kg)"].mean()
```

```
#Calculando a média de todos os valores numéricos do dataframe

dados.mean()
```
